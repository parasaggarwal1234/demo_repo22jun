#### List and Dict Comprehension

```python
r = [i if i%2==0 else (i*3 if i%5==0 else i+5) for i in range(2,30,3)]
print(r)
# i=2,5,8,11,14,17,20,23,26,29
# i%2==0 => i  => 2,8,14,20,26,
# i%5==0 => i*3 => 15,
# else => i+5  => 16,22,28,34
```

#### map,filter, reduce

map(function,iterable)<br>
filter(function,iterable)<br>
reduce(function,iterable)<br>

```python
from functools import reduce
```

```python
r = [10,15,20,25,30,35,40,45]

r1 = list(map(lambda a : a*2+5, r))
r2 = list(filter(lambda a : a%3==0, r))
print(r1)
print(r2)
```

```python
r = [10,15,20,25,30,35,40,45]
r3 = reduce(lambda a,b : a+b, r)
print(r3)
# S1 => a=10,b=15 => a+b = 25
# S2 => a=25,b=20 => a+b =45
# S3 => a=45,b=25 => a+b =70
# S4 => a=70,b=30=> a+b = 100
# S5 => a=100,b=35 => a+b =135
# S6 => a=135,b=40 => a+b =175
# S7 => a=175,b=45 => a+b =220
```

```python
r = [10,15,20,25,30,35,40,45,50.5]

r1 = list(map(lambda a : a*2+5, r))
r2 = list(filter(lambda a : a%3==0, r))
print(r1)
print(r2)
```

```python
r = ['Welcome','true','India','PGA21','laptop']
r1 = list(map(lambda x : x.lower(),r))
r2 = list(map(lambda x : len(x),r))
print(r1)
print(r2)
```

```python
r = [10,15,20,25,30,35,40,45]

r1 = list(filter(lambda a : a*2+5, r))
r2 = list(map(lambda a : a%3==0, r))
print(r1)
print(r2)
```

#### sorted()

```python
w = [12,14,10,9,15,13]
r1 = sorted(w)               # ascedning order
r2 =  sorted(w,reverse=True)  # descending order
print(r1)
print(type(r1))
print(r2)
```

```python
d = {'a1':10,'a3':8,'a4':15,'a2':25}
d1 = sorted(d.keys())               # ascending order
d2 = sorted(d.keys(),reverse=True)  # descending order
d3 = sorted(d.values())             # ascending order
d4 = sorted(d.values(),reverse=True) # descending order
print(d1)
print(d2)
print(d3)
print(d4)
```

```python
d = {'a1':10,'a3':8,'a4':15,'a2':25}
r1 = sorted(d.items(), key=lambda x: x[1])  # ('a1',10),('a3',8),('a4',15),('a2',25)
# sorted in increasing order of values
print(r1)
r2 = sorted(d.items(), key=lambda x: x[1], reverse=True) 
# sorted in increasing order of values
print(r2)
r3 = dict(sorted(d.items(), key=lambda x: x[1]))
# sorted in increasing order of values
print(r3)
```

### Patterns

```python
# i)
n = int(input('Enter a number: '))
for i in range(1,n+1):
    for j in range(1,n+1):
        if i==j or i+j==(n+1):
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
# j)
n = int(input('Enter a number: '))
for i in range(n,0,-1):
    for j in range(1,n+1):
        if i==j or i+j==(n+1):
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
# k)
n = int(input('Enter a number: '))
for i in range(1,n+1):
    for j in range(1,n+1):
        if (i+j)%2==0:
            print(1,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
# l)
n = int(input('Enter a number: '))
for i in range(1,n+1):
    for j in range(1,n+1):
        print(i+j-1,end=" ")
    print()
```

### Pivot table

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.columns = ['Gender','Race_Eth','PLEC','Lunch','TPEC','MS','RS','WS']
df.head()
```

```python
df.rename(columns={'Lunch':'Meal'},inplace=True)
df.head()
```

#### Pivot table
Default aggegate function is mean

```python
d1 = pd.pivot_table(data=df,index='Race_Eth',columns=['TPEC','Gender'],values=['MS','RS'])
d1.head()
```

```python
d1 = pd.pivot_table(data=df,index=['Race_Eth','PLEC'],columns=['TPEC','Gender'],values=['RS','WS'])
d1
```

```python
d3 = pd.pivot_table(data=df,index=['Race_Eth','PLEC'],columns=['TPEC','Gender'],
                    values=['RS','WS'], aggfunc=sum)
d3
```

```python
d3 = pd.pivot_table(data=df,index=['Race_Eth','PLEC'],columns=['TPEC','Gender'],
                    values=['RS','WS'], aggfunc=sum, fill_value=0)
d3
```

### Stack and Unstack
