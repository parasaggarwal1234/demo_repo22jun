### Remove null values

### 1) df.dropna()
Parameters<br>
1) how= 'any' : If any NA values are present, drop that row or column depending on the axis<br>
2) how = 'all' : If all values are NA, drop that row or column depending on the axis


```python
d = {'Name':['Alisa','Bobby','Cathrine',np.nan,'Rocky','Sebastian','Jaqluine',np.nan,'David'],
     'Score1':[62,47,np.nan,74,31,77,np.nan,63,42],
     'Score2':[89,87,67,np.nan,47,72,76,79,44],
     'Score3':[56,np.nan,77,45,73,62,74,np.nan,71]}
df1 = pd.DataFrame(d)
df1
```

```python
d2 = df1.dropna(axis=0, how ='any') # axis =0 is row wise
d2
```

```python
d3 = df1.dropna(axis=1, how ='any')  # axis =1 is column wise
d3
```

```python
df7 = pd.DataFrame({"Name": ['Flash', 'Batman', 'Superman','Hulk',np.nan],
                   "Power": [np.nan, 'Rich', 'Strength',np.nan,np.nan],
                    "color": ['red',np.nan,'RedBlue','Green',np.nan]
                  })
df7
```

```python
# Drop the rows where at least one element is missing.
df7.dropna()  # default axis=0, how='any'
```

```python
# Drop the columns where at least one element is missing.
df7.dropna(axis=1)  # how='any'
# OR
# df7.dropna(axis='columns')
```

```python
df7
```

```python
# Drop the rows where all elements are missing.
df7.dropna(how='all')  # axis=0, row wise
```

```python
# Drop the columns where all elements are missing.
df7.dropna(how='all',axis=1) # axis=1, column wise
```

### Apply, ApplyMap and Map Functions

<b>apply():</b> 
<ol>
<li>Apply is defined on both series and dataframe</li>
<li>It operates row or column wise, i.e. along an axis. </li>
<li>Suited for more complex operations and aggregations(sum,mean)</li>
<li>Accepts a function as its agrument</li>
</ol>
<b>applymap():</b>
<ol>
<li>applymap is defined on DataFrames only</li>
<li>It operates element wise on dataframes</li>
<li>Accepts a function as its argument</li>    
</ol>
<b>map():</b>
<ol>
<li>map is defined on Series only</li>
<li>map operates element wise on Series</li>
<li>map accepts dict. It may also accept a function as its argument</li>    
</ol>
</ol>
<br>

#### applymap()

```python
d1f = pd.DataFrame([[10, 15,73], [25, 40, 84]])
d1f.columns=['A','B','C']
d1f
```

```python
def fun1(val):
    if(val%2==0):
        return val+10
    else:
        return val-5
```

```python
d1f.applymap(fun1)
```

```python
d1f
```

```python
d2f = d1f.applymap(lambda x:x**2)
d2f
```

```python
d3 = [['Namaste','Welcome','Adios'],['Bye','See you soon','Good Morning']]
d2f = pd.DataFrame(d3,columns=['G1','G2','G3'])
d2f
```

```python
d3f = d2f.applymap(lambda x: len(x))
d3f
```

#### map()

```python
d1 = {'Country':['China','India','USA','Indonesia','Brazil'],
      'Population':[1403500365,1324171354,322179605,261115456,                                                                                     207652865]}
df= pd.DataFrame(d1)
df
```

```python
country_capital = {'Germany':'Berlin','Brazil':'Braslia','Budapest':'Hungary',
                  'China':'Beijing','India':'New Delhi','Norway':'Oslo',
                  'France':'Paris','USA':'Washington'
                }
# 'Indonesia': 'Jakarta'
country_capital
```

```python
df
```

```python
country_capital
```

```python
df['Capital'] = df['Country'].map(country_capital)
df.head()
```

```python
df['Population_in_Mil'] = df['Population'].map(lambda x : round((x/1000000),2))
df.head()
```
