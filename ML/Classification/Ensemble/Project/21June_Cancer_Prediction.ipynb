#### Handling Outliers

```python
def outlier_handling(x):
    x = x.clip(upper = x.quantile(0.98))
    return x
```

```python
print(num_cols)
```

#### Handling MultiCollinearity

```python
corr.iloc[1,2]
# 0.3403508450500664
# corr
```

```python
len(corr.columns)
```

```python
def high_corr(data, thresh):
    res = set()
    corr = data.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if corr.iloc[i,j] >= thresh:
                res.add(corr.columns[i])
    return list(res)

multicoll_cols = high_corr(df,0.9)
print(multicoll_cols)
print(len(multicoll_cols))
```

```python
df1 = df.copy()
df1.head(3)
```

```python
df1.drop(multicoll_cols,axis=1,inplace=True)
print(len(df1.columns))
print(df1.columns)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
df1['diagnosis'] = lb.fit_transform(df1['diagnosis'])
```

```python
df1['diagnosis'].value_counts()
```

```python
df1.dtypes.value_counts()
```

```python
x = df1.drop('diagnosis',axis=1)
y = df1['diagnosis']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,
                                                 random_state=15)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
y_test.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_re-port,accuracy_score
```

```python
m1 = DecisionTreeClassifier(criterion='entropy',max_depth=10,min_samples_split=13)
m1.fit(x_train,y_train)
print('Train_Score',m1.score(x_train,y_train))
print('Test_Score',m1.score(x_test,y_test))
```

```python
ypred_m1 = m1.predict(x_test)
print(ypred_m1)
```

```python
cm = confusion_matrix(y_test,ypred_m1)
print(cm)
print(classification_report(y_test,ypred_m1))
```

#### 3) AdaBoost Classfier

```python
def gen_metrics(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    y_pred = model.predict(x_test)
    print('Predictions\n',y_pred)
    acc = accuracy_score(y_test,y_pred)
    print('Training score',train_score)
    print('Testing score',test_score)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
dt1 = DecisionTreeClassifier(max_depth=2)
m4 = AdaBoostClassifier(dt1,n_estimators=60,learning_rate=1)
gen_metrics(m4,x_train,x_test,y_train,y_test)
```
