```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
x = iris.data[:,:2] # independent features - sepal length,sepal width
y = iris.target     # dependent data - 3 categories
```

```python
print(x.shape)
print(y.shape)
```

```python
print(type(x))
print(type(y))
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
from sklearn.svm import SVC, SVR
```

### Plot SVM with different kernels

```python
# kernel - 'linear','rbf','poly','sigmoid'
# C - 
# degree - 
# gamma - rbf, poly
```

```python
kernel = ['linear','rbf','poly']
for i in kernel:
    model = SVC(kernel=i)       # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with kernel {i}')
    plt.show()
```

### Plot Rbf with different Kernels

```python
g = [0.01,0.1,1,10,100]
for i in g:
    model = SVC(kernel='rbf',gamma=i)  # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with rbf kernel  with gamma={i}')
    plt.show()
```

### Linear Kernel with different C

```python
c = [0.0001,0.001,0.01,0.1,1,10,100]
for i in c:
    model = SVC(kernel='linear',C=i)             # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with poly kernel with C={i}')
    plt.show()
```

####  Poly kernel with different C

```python
c = [0.01,0.1,1,5,10,30,60,100]
for i in c:
    model = SVC(kernel='poly',C=i)          # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with linear kernel with C={i}')
    plt.show()
```

### Plot poly kernel with different degree¶

```python
deg = [1,2,3,4,5]
for i in deg:
    model = SVC(kernel='poly',degree=i)          # default C=1.0
    model.fit(x,y)
    plot_decision_regions(x, y, clf=model)
    plt.xlabel('Sepal Length')
    plt.ylabel('Sepal Width')
    plt.title(f'SVM with poly kernel with degree={i}')
    plt.show()
```
