### Plotting a Normal Distribution or Gaussian Distribution

```python
a = np.linspace(1,100,200)

def generate_norm_data(a):
    m = np.mean(a)
    s = np.std(a)
    gauss = 1/(s*np.sqrt(2*np.pi)) *(np.exp(-0.5*((a-m)/s)**2))
    return gauss
                 
        
r = generate_norm_data(a)
print(r)
```

```python
plt.plot(a,r)
plt.grid()
plt.show()
```

Q1) For the given data, create a function to find its skewness and adjusted skewness<br>
data  = [2.5, 3.7, 6.6, 9.1, 9.5, 10.7, 11.9, 21.5, 22.6, 25.2]

```python
data = [2.5, 3.7, 6.6, 9.1, 9.5, 10.7, 11.9, 21.5, 22.6, 25.2]

def skewness(data):
    m = np.mean(data)
    m3 = sum([(i-m)**3 for i in data])/len(data)
    m2 = sum([(i-m)**2 for i in data])/len(data)
    skew = (m3)/(m2)**(3/2)
    print('Skewness',skew)
    adj_skew = np.sqrt(len(data)*(len(data)-1))/(len(data)-2)*skew
    print('Adj_skew',adj_skew)
    
skewness(data)
```

### Kurtosis

1) In statistics, kurtosis is a measure of relative peakedness of a probability distribution, or alternatively how heavy or how light its tails are. A value of kurtosis describes how different the tails of a given probability distribution are from a normal distribution.<br>
2) It indicates whether the tail of distribution extends beyond the ±3 standard deviation of the mean or not.<br>
3) Kurtosis measures whether there are extreme values in either of the tails

<img src="kurt_img.png">

There are 3 types of Kurtosis:
1) <b>Positive excess kurtosis ((k-3)>0) </b> When excess kurtosis, given by (kurtosis – 3), is positive, then the distribution has a sharp peak and is called a leptokurtic distribution.<br> 
2) <b>Negative excess kurtosis ((k-3)<0)  </b> When excess kurtosis, given by (kurtosis – 3), is negative, then the distribution has a flat peak and is called a platykurtic distribution.<br>
3) <b>Zero excess kurtosis  ((k-3)=0) </b> – When excess kurtosis, given by (kurtosis – 3), is zero, then the distribution follows a normal distribution and is also called a mesokurtic distribution.<br>

d1 = [26, 12, 16, 56, 112, 24]

```python
# w = [2,4,6,8,10]
# m = np.mean(w)
# print(m)
# m4 = [(i-m)**4 for i in w]
# print('m4',m4,sum(m4))
# m2 = [(i-m)**2 for i in w]
# print('m4',m2,sum(m2))
```

```python
d1 = [26, 12, 16, 56, 112, 24]

def kurtosis(x):
    m = np.mean(x)
    m4 = sum([(i-m)**4 for i in x])/len(x)
    m2 = sum([(i-m)**2 for i in x])/len(x)
    kurt = m4/(m2**2)
    print('Kurtosis',kurt)
    
kurtosis(d1)
```

```python
d1 = [26, 12, 16, 56, 112, 24]

print(kurtosis(d1,fisher=False))
print(kurtosis(d1,bias=True))
```

### Central Limit Theorem
1) CLT states that if you have a population with mean μ, sd σ, and take sufficiently large random samples from the population with replacement, then the distribution of the sample means will be approximately be normally distributed.

2) The sample mean will approximately be normally distributed for large sample sizes, regardless of the distribution from which we are sampling. 
 

Ex:
Central Limit Theorem suggests that if you randomly draw a sample of your customers, say 1000 customers, this sample itself might not be normally distributed. But if you now repeat the experiment say 100 times, then the 100 means of those 100 samples (of 1000 customers) will make up a normal distribution.

### Confidence Interval and Confidence Level

Often in statistics we’re interested in measuring population parameters – numbers that describe some characteristic of an entire population. Two of the most common population parameters are:<br>
1) Population Mean<br>
2) Population Parameter

### Confidence Interval
1) A confidence interval is a range of values that is likely to contain a population parameter with a certain level of confidence.<br>
2) Confidence intervals measure the degree of uncertainty or certainty in a sampling method. They are most often constructed using confidence levels of 95% or 99%.or 90%<br>
3) A confidence interval displays the probability that a parameter will fall between a pair of values around the mean.<br>
4) It is range of values consisting of lower and upper bound<br>


### Cofidence Level
1) The confidence level tells you how sure you can be. It is expressed as a percentage and represents how often the true percentage of the population who would pick an answer that lies within the confidence interval. 

#### Ex1
The 95% confidence level means you can be 95% certain; the 99% confidence level means you can be 99% certain. Most researchers work for a 95% confidence level.
When you put the confidence level and the confidence interval together, you can say that you are 95% sure that the true percentage of the population is between 43% and 51%.

#### Ex2
A baseball coach was curious about the true mean speed of fastball pitches in his league. The coach recorded the speed in kilometers per hour of each fastball in a random sample of 100 pitches and constructed a 95%, percent confidence interval for the mean speed. The resulting interval was (110,120).
